```python
r, g, b = imagen.split()
```

```python
r1 = r.point(lambda i: i * 1) #rojo
g1 = g.point(lambda i: i * 0)
b1 = b.point(lambda i: i * 0)
```

```python
r2 = r.point(lambda i: i * 0.1) 
g2 = g.point(lambda i: i * 0.1)
b2 = b.point(lambda i: i * 1) #azul
```

```python
r3 = r.point(lambda i: i * 0)
g3 = g.point(lambda i: i * 1)
b3 = b.point(lambda i: i * 0) #verde
```

```python
r4 = r.point(lambda i: i * 1)
g4 = g.point(lambda i: i * 1)
b4 = b.point(lambda i: i * 0) #amarillento
```

```python
r5 = r.point(lambda i: i * 1)
g5 = g.point(lambda i: i * 0)
b5 = b.point(lambda i: i * 0.8) #fucsia
```

```python
resultado1 = Image.merge('RGB', (r1, g1, b1))
resultado2 = Image.merge('RGB', (r2, g2, b2))
resultado3 = Image.merge('RGB', (r3, g3, b3))
resultado4 = Image.merge('RGB', (r4, g4, b4))
resultado5 = Image.merge('RGB', (r5, g5, b5))

resultado1.save('pruebarojo.jpg')

print (resultado1.size)
```
